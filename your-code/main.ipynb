```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset
url = "https://archive.ics.uci.edu/ml/machine-learning-databases/00292/Wholesale%20customers%20data.csv"
data = pd.read_csv('https://archive.ics.uci.edu/ml/machine-learning-databases/00292/Wholesale%20customers%20data.csv')

# === 1. Overview ===
print("\n--- First 5 rows ---")
print(data.head())

print("\n--- Column Info ---")
print(data.info())

print("\n--- Descriptive Statistics ---")
print(data.describe())

#  2. What does each column mean? 
print("\n--- Column meanings ---")
column_info = {
    'Channel': '1=Horeca (Hotel/Restaurant/Cafe), 2=Retail',
    'Region': '1=Lisbon, 2=Oporto, 3=Other Region',
    'Fresh': 'Annual spending on fresh products',
    'Milk': 'Annual spending on milk products',
    'Grocery': 'Annual spending on grocery products',
    'Frozen': 'Annual spending on frozen products',
    'Detergents_Paper': 'Annual spending on detergents and paper products',
    'Delicassen': 'Annual spending on delicatessen products'
}
for col, desc in column_info.items():
    print(f"{col}: {desc}")

# 3. Check for missing data 
print("\n--- Missing Values ---")
print(data.isnull().sum())

#  4. Categorical columns
# Convert 'Channel' and 'Region' to category type
data['Channel'] = data['Channel'].astype('category')
data['Region'] = data['Region'].astype('category')

#  5. Correlation Matrix 
print("\n--- Correlation Matrix ---")
corr = data.drop(columns=['Channel', 'Region']).corr()
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title("Correlation Matrix")
plt.show()

#  6. Pairplot 
sns.pairplot(data.drop(columns=['Channel', 'Region']))
plt.show()

# 7. Boxplots (Outlier detection) 
for col in data.columns[2:]:  # Skip Region and Channel
    plt.figure(figsize=(6, 1.5))
    sns.boxplot(x=data[col])
    plt.title(f'Boxplot of {col}')
    plt.show()

#  8. Histograms (Distribution)
data.drop(columns=['Region', 'Channel']).hist(bins=20, figsize=(12, 8))
plt.suptitle("Histograms of Numerical Columns")
plt.show()

#  9. Skewness
print("\n--- Skewness ---")
print(data.drop(columns=['Region', 'Channel']).skew())

# 10. Check for Pareto Principle (80/20 Rule) 
# Check for 'Grocery' spending
sorted_grocery = data['Grocery'].sort_values(ascending=False).reset_index(drop=True)
cumulative = sorted_grocery.cumsum() / sorted_grocery.sum()

plt.plot(np.arange(len(cumulative)) / len(cumulative), cumulative)
plt.xlabel('Fraction of Customers')
plt.ylabel('Cumulative Grocery Spending')
plt.title('Pareto Principle Check for Grocery')
plt.grid(True)
plt.axvline(0.2, color='r', linestyle='--', label='20% Customers')
plt.axhline(0.8, color='g', linestyle='--', label='80% Spending')
plt.legend()
plt.show()

# How many customers contribute to 80% of Grocery sales?
top_20_pct = cumulative[cumulative < 0.8].count() / len(cumulative)
print(f"\nPercentage of customers contributing to 80% Grocery spend: {top_20_pct:.2%}")
```

# Challenge 3 - Data Preprocessing

One problem with the dataset is the value ranges are remarkably different across various categories (e.g. `Fresh` and `Grocery` compared to `Detergents_Paper` and `Delicassen`). If you made this observation in the first challenge, you've done a great job! Keep on the good work!

Diverse value ranges in different features could cause issues in our clustering. The way to reduce the problem is through feature scaling. We'll use this technique again with this dataset.

#### We will use the `StandardScaler` from `sklearn.preprocessing` and scale our data. Read more about `StandardScaler` [here](https://scikit-learn.org/stable/modules/generated/sklearn.preprocessing.StandardScaler.html#sklearn.preprocessing.StandardScaler).

*After scaling your data, assign the transformed data to a new variable `customers_scale`.*

Scaled already above!

```python
from sklearn.preprocessing import StandardScaler
import numpy as np
import pandas as pd

# Assume `data` is already loaded
# Step 1: Drop non-numeric (categorical) features
df_numeric = data.drop(['Region', 'Channel'], axis=1)

# Step 2: Apply log transform to reduce skew 
df_log = np.log1p(df_numeric)  # log1p avoids log(0) issues

# Step 3: Scale the data using StandardScaler
scaler = StandardScaler()
customers_scale = scaler.fit_transform(df_log)

# (Optional) Convert to DataFrame for inspection
customers_scale = pd.DataFrame(customers_scale, columns=df_numeric.columns)

# Preview the scaled data
print("Data scaled and stored in 'customers_scale'")
print(customers_scale.head())

```

# Challenge 4 - Data Clustering with K-Means

Now let's cluster the data with K-Means first. Initiate the K-Means model, then fit your scaled data. In the data returned from the `.fit` method, there is an attribute called `labels_` which is the cluster number assigned to each data record. What you can do is to assign these labels back to `customers` in a new column called `customers['labels']`. Then you'll see the cluster results of the original data.

```python

from sklearn.cluster import KMeans

# Step 1: Initialize the model with a chosen number of clusters
kmeans = KMeans(n_clusters=3, init='k-means++', random_state=42)

# Step 2: Fit the model to the scaled data
kmeans.fit(customers_scale)

# Step 3: Get cluster labels
labels = kmeans.labels_

# Step 4: Assign labels to original data
customers = data.copy()  # make a copy to preserve original data
customers['labels'] = labels

# Preview
print("Clustering complete. 'labels' column added to customers:")
print(customers.head())
```

```python
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

wcss = []

# Cluster numbers from 1 to 10
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, random_state=42)
    kmeans.fit(customers_scale)
    wcss.append(kmeans.inertia_)  # inertia_ = WCSS

# Plot the elbow curve
plt.figure(figsize=(8,5))
plt.plot(range(1, 11), wcss, marker='o')
plt.title('Elbow Method For Optimal k')
plt.xlabel('Number of clusters (k)')
plt.ylabel('WCSS')
plt.xticks(range(1,11))
plt.show()
```

Columns to visualize:

# Bonus Challenge 2 - Changing K-Means Number of Clusters

As we mentioned earlier, we don't need to worry about the number of clusters with DBSCAN because it automatically decides that based on the parameters we send to it. But with K-Means, we have to supply the `n_clusters` param (if you don't supply `n_clusters`, the algorithm will use `8` by default). You need to know that the optimal number of clusters differs case by case based on the dataset. K-Means can perform badly if the wrong number of clusters is used.

In advanced machine learning, data scientists try different numbers of clusters and evaluate the results with statistical measures (read [here](https://en.wikipedia.org/wiki/Cluster_analysis#External_evaluation)). We are not using statistical measures today but we'll use our eyes instead. In the cells below, experiment with different number of clusters and visualize with scatter plots. What number of clusters seems to work best for K-Means?

```python
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans

# Define a plotting function
def plot_clusters(n_clusters, x='Grocery', y='Milk'):
    kmeans = KMeans(n_clusters=n_clusters, random_state=42)
    labels = kmeans.fit_predict(customers_scale)

    plt.figure(figsize=(6, 5))
    sns.scatterplot(x=customers[x], y=customers[y], hue=labels, palette='tab10', alpha=0.7)
    plt.title(f'K-Means Clustering with {n_clusters} Clusters')
    plt.xlabel(x)
    plt.ylabel(y)
    plt.legend(title='Cluster')
    plt.show()

# Try different numbers of clusters
for k in [2, 3, 4, 5, 6]:
    plot_clusters(k)
```

# Bonus Challenge 3 - Changing DBSCAN `eps` and `min_samples`

Experiment changing the `eps` and `min_samples` params for DBSCAN. See how the results differ with scatter plot visualization.

```python
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import DBSCAN

# Function to apply DBSCAN and plot results
def plot_dbscan(eps, min_samples, x='Grocery', y='Milk'):
    dbscan = DBSCAN(eps=eps, min_samples=min_samples)
    labels = dbscan.fit_predict(customers_scale)

    plt.figure(figsize=(6, 5))
    sns.scatterplot(x=customers[x], y=customers[y], hue=labels, palette='tab10', alpha=0.7)
    plt.title(f'DBSCAN: eps={eps}, min_samples={min_samples}')
    plt.xlabel(x)
    plt.ylabel(y)
    plt.legend(title='Cluster')
    plt.show()

# Try different DBSCAN settings
params_to_try = [
    (0.3, 3),
    (0.5, 5),   # default
    (0.7, 5),
    (0.5, 10),
    (0.8, 10)
]

for eps, min_samples in params_to_try:
    plot_dbscan(eps, min_samples)
```
